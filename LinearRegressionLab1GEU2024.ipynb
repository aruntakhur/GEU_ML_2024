Error function for 1 degree polynomial with zero intercept
Equation of the line is ​y = mx + c
1. Slope of the line is 1 i.e m = 1.
2. Intercept is 0 i.e. c = 0.

```python
import numpy as np
import matplotlib.pyplot as plt
m = 1
x_data = np.linspace(1,10,11)
y_data = m*x_data
plt.title('Graph of equation y = x')
plt.scatter(x_data,y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Plotting Error function
Equation of error function is:

 $Error = ∑^n_{i=1}(m*x_i - p_i)  //comment: m*x_i=y_i $

Above equation has only 1 parameter that is w1.

```python
m_diff_slopes = np.linspace(-6,9,50)
error = 0
it = len(x_data)
for i in range(it):
  error += (m_diff_slopes*x_data[i]- y_data[i])**2
plt.plot(m_diff_slopes,error)
plt.xlabel('W')
plt.ylabel('Error')
plt.title('W vs Error')
plt.scatter(m_diff_slopes,error,c='red')
plt.show()
```

From the above graph we can see that minimum value of Error is 0 when w is equivalent to 1

# Plot of line which minimized the error with the given trianing dataset.

```python
m = 1
x_new = np.linspace(1,10,25)
p = m*x_new
plt.scatter(x_new,p,c='red')
plt.scatter(x_data,y_data)
plt.title('Graph of equation y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Using sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
#X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
#y = np.dot(X, np.array([1, 2])) + 3
x_data = x_data.reshape(-1, 1)
y_data = y_data.reshape(-1, 1)
reg = LinearRegression().fit(x_data, y_data)
reg.score(x_data, y_data)
print(reg.coef_)
print(reg.intercept_)
p = reg.coef_*x_new + reg.intercept_
plt.scatter(x_data,y_data)
plt.scatter(x_new,p,c='red')
plt.title('Graph of equation y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
